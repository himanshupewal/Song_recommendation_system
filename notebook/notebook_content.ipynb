```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd
```

```python
df_colab_filtering.columns
```

```python
frequency_encode_cols = ["year"]
ohe_cols = ["artist","time_signature","key"]
tfidf_col = "tags"
standard_scale_cols = ["duration_ms","loudness","tempo"]
min_max_scale_cols = ["acousticness","danceability","energy","instrumentalness","liveness","speechiness","valence"] 
```

```python
class CountEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.count_maps_ = {}

    def fit(self, X, y=None):
        for col in X.columns:
            self.count_maps_[col] = X[col].value_counts()
        return self

    def transform(self, X):
        X_encoded = X.copy()
        for col in X.columns:
            X_encoded[col] = X_encoded[col].map(self.count_maps_[col]).fillna(0)
        return X_encoded
```

```python


transformer = ColumnTransformer(transformers=[
    ("count_encode", CountEncoder(), frequency_encode_cols),
    ("ohe", OneHotEncoder(handle_unknown="ignore"), ohe_cols),
    ("tfidf", make_pipeline(TfidfVectorizer(max_features=85)), tfidf_col),
    ("standard_scale", StandardScaler(), standard_scale_cols),
    ("min_max_scale", MinMaxScaler(), min_max_scale_cols)
], remainder='passthrough', n_jobs=-1)
```

```python
def recommend_songs(song_name, songs_data, transformed_data,k=10):
    song_input = df_songs[df_songs["name"] == song_name]
    input_vector = transformer.transform(song_input)
    similarity_scores = cosine_similarity(transformer.transform(df_colab_filtering), input_vector)
    top_10_songs_indexes = np.argsort(similarity_scores.ravel())[-11:]
    top_10_songs_names = df_songs.iloc[top_10_songs_indexes]
    return top_10_songs_names  
```

```python
def recommend_songs(song_name, songs_data, transformed_data,k=10):
    song_row = songs_data.loc[songs_data['name']==song_name,:]
    if song_row.empty:
        print(f"Song '{song_name}' not found in the dataset.")
        
    else:
        song_index = song_row.index[0]
        print(f"Song '{song_name}' found at index {song_index}.")

        input_vector = transformed_data[song_index].reshape(1, -1)
        similarity_scores = cosine_similarity( input_vector,transformed_data)
        print(f"Similarity scores calculated for song '{song_name}'.")
        top_k_indices = np.argsort(similarity_scores.ravel())[-k-1:-1][::-1]
        print(f"Top {k} similar songs found for '{song_name}'.")
        top_k_songs_names= songs_data.iloc[top_k_indices,:]

        top_k_list = top_k_songs_names[["name","artist","spotify_preview_url"]].reset_index(drop=True)
        return top_k_list 
         
```

```python
recommend_songs("Whenever, Wherever", df_songs, transformed_df, k=10)
```
